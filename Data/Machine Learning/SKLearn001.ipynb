```python
# This model is to predict the prices of airbnb rental based on number of reviews

# setting up the Pandas
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame
lsd = pd.read_csv('numreview.csv')
lsd = lsd.loc[lsd['price']<2000]
lsd = lsd.loc[lsd['price']>1]
lsd
```

```python
#reset the values into the model
y = lsd.price.values.reshape(-1, 1)
x = lsd.number_of_reviews.values.reshape(-1, 1)

print("Shape: ", x.shape, y.shape)
y
```

```python
# show the results in a scatter plot
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

### BEGIN SOLUTION

model = LinearRegression()

### END SOLUTION
```

```python
model.fit(x, y)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
x_min = np.array([[x.min()]])
x_max = np.array([[x.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.scatter(x, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
```
