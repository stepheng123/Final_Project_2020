```python
# This model is to predict the prices of airbnb rental based on the location in terms of latitude

# setting up the Pandas
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ausexp = pd.read_csv('aus_export.csv')
ausexp = ausexp.loc[ausexp['Year']>1990]
ausexp.head()
```

```python
x = ausexp.Year.values.reshape(-1, 1)
y = ausexp.Value.values.reshape(-1, 1)

print("Shape: ", x.shape, y.shape)
x
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

### BEGIN SOLUTION

model = LinearRegression()

### END SOLUTION
```

```python
model.fit(x, y)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
x_min = np.array([[x.min()]])
x_max = np.array([[x.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.scatter(x, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
```

```python
model.predict(np.array([[2021]]))
```

```python
years = [2020, 2021, 2022, 2023, 2024, 2025, 2030, 2040]
results = []
for i in years:
    print(model.predict(np.array([[i]])))
    results.append(model.predict(np.array([[i]])))
```
